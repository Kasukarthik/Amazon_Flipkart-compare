```python
# separate the product_name and retail_price from the original data. 
amazon=amazon_data[['product_name','retail_price']]
flipkart=flipkart_data[['product_name','retail_price']]
```

```python
# compare the amazon and flipkart data 
similar_products=amazon.compare(flipkart).dropna()
```

```python
# check any missing values present in the data
similar_products.isnull().sum()
```

```python
# rename the columns to our convenience 
similar_products.columns='amazon_product','flipkart_product','amazon_retail','flipkart_retail'
```
